```python
import pandas as pd
import numpy as np
import sklearn
from sklearn.decomposition import PCA
from pandas import DataFrame
import matplotlib.pyplot as plt
from datetime import datetime as dt
from sklearn.preprocessing import StandardScaler
```

```python
with pd.HDFStore('AMPds2.h5') as hdf:
    # This prints a list of all group names:
    print(hdf.keys())
```

```python
df = pd.read_hdf("AMPds2.h5", key='/building1/elec/meter9')
df.head()
```

```python
days_meters=pd.DataFrame()

for i in range(1,22):
    print(i)
    key1 = "/building1/elec/meter"+str(i)
    df = pd.read_hdf("AMPds2.h5", key=key1)
    # Convert Apparent and Reactive --> Real Power to allow aggregation and yield relevant info. 
    df[("meter"+str(i),"")] = (df[("power", "apparent")]**2 - df[("power", "reactive")]**2)**.5
    df['time'] = df.index
    to_add = df.groupby(pd.to_datetime(df.time).dt.date).agg({("meter"+str(i),""): 'sum'})
    if days_meters.empty:
        days_meters = to_add / 60
    else: 
        to_add = to_add / 60
        days_meters = pd.merge(days_meters, to_add, left_index=True, right_index=True)
        
```

```python
sklearn.preprocessing.normalize(days_meters, norm='l2', axis=1, copy=False, return_norm=False)
```

```python
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(days_meters)
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['principal component 1', 'principal component 2'])
finalDf = pd.concat([principalDf], axis = 1)
# finalDf.columns.values[2] = "target"
# targets=np.unique(dboolmeta.iloc[:,2])
```

```python
finalDf.index = days_meters.index
finalDf["full_date"] = pd.to_datetime(finalDf.index)
finalDf["date"] = finalDf.full_date.dt.strftime("%m-%d")
finalDf["month"] = finalDf.full_date.dt.strftime("%m")
```

```python
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('PC2 v PC1', fontsize = 20)
ax.scatter(finalDf['principal component 1']
            , finalDf['principal component 2']
            , alpha = 0.15
            , s = 20)
ax.grid()
```

```python
targets = np.unique(finalDf["month"])
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('PC2 v PC1, colored by population', fontsize = 20)
colors = []
for i in range(1,13):
    colors.append((i / 12, i / 12, i / 12))
for target, color in zip(targets,colors):
    indicesToKeep = finalDf["month"] == target
    ax.scatter(finalDf.loc[indicesToKeep, 'principal component 1']
               , finalDf.loc[indicesToKeep, 'principal component 2']
               , c = [color]
               , alpha = 0.15
               , s = 20)
ax.legend(targets)
ax.grid()
```

```python
targets = np.unique(finalDf["date"])
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('PC2 v PC1, colored by population', fontsize = 20)
colors = []
for i in range(1,366):
    colors.append((i / 365, i / 365, i / 365))
for target, color in zip(targets,colors):
    indicesToKeep = finalDf["date"] == target
    ax.scatter(finalDf.loc[indicesToKeep, 'principal component 1']
               , finalDf.loc[indicesToKeep, 'principal component 2']
               , c = [color]
               , alpha=0.15
               , s = 20)
ax.grid()
```

```python
finalDf["day"] = finalDf.full_date.dt.strftime('%A')
```

```python
targets = np.unique(finalDf["day"])
print(targets)
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('PC2 v PC1, colored by weekend', fontsize = 20)
colors = ['r','c','k','k','c','c','c']

for target, color in zip(targets,colors):
    indicesToKeep = finalDf["day"] == target
    ax.scatter(finalDf.loc[indicesToKeep, 'principal component 1']
               , finalDf.loc[indicesToKeep, 'principal component 2']
               , c = [color]
               , alpha=0.35
               , s = 20)
ax.grid()
```

```python
X_std = StandardScaler().fit_transform(days_meters)

pca2 = PCA(n_components = 16)
pca2.fit(X_std)
plt.plot((pca2.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('explained variance')
plt.show()
```
